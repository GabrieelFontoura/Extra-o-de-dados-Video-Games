```python
videogames = pd.read_csv('vgsales.csv')

videogames.head()
```

```python
# Ira retornar um resumo de todas as variaveis ou colunas da tabela NUMERICAS
videogames.describe()
```

```python
# Tipo dos dados
videogames.dtypes
```

```python
#Renomeando colunas
videogames.columns = ['Ranking', 'Nome', 'Plataforma', 'Ano',
'Gênero', 'Editora', 'Vendas América do Norte', 'Vendas EUA','Vendas Japão'
, 'Outras vendas', 'Vendas Global']

videogames.head()
```

```python
# Visualizando valores nulos
videogames[videogames['Ano'].isnull()].head()
```

```python
# Quantos jogos foram lançados por plataforma: 
videogames['Plataforma'].value_counts()
```

```python
# Grafico horizontal

# Quantos jogos foram lançados por cada plataforma, agora em um grafico:
# Função .plot CRIA o GRAFICO
titulos_lancados = videogames['Plataforma'].value_counts().head(10).plot(kind='bar', figsize=(11,5), grid = False, rot=0, color='green')

# Melhorando o grafico a baixo
plt.title('Os 10 Video-Games com mais titulos já lançados') # Titulo do grafico
plt.xlabel('Video-Games') # Nomeando o eixo X, onde fica o nome dos video games
plt.ylabel('Quantidade de jogos lançados') # Nomeando o eixo Y, onde fica a quantidade de jogos
plt.show() # Exibindo o grafico
```

```python
# Grafico horizontal

top_10_vendidos = videogames[['Nome', 'Vendas Global']].head(10).set_index('Nome').sort_values('Vendas Global', ascending=True)
top_10_vendidos.plot(kind='barh',figsize=(11, 7), grid=False, color = 'darkred', legend=False)

plt.title('Os 10 jogos mais vendidos no mundo')
plt.xlabel('Total de vendas em milhoes de dolares')
plt.show
```
